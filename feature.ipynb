```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
help(sns.pairplot)
```

```python
def splom_viz(df, labels=None):
    ax = sns.pairplot(df, hue=labels, diag_kind='hist', size=1)
    plt.savefig("fea_pairplot.svg")
    plt.savefig("fea_pairplot.png")
    plt.show()
```

```python
df = pd.DataFrame(np.hstack((X.values, Y.values)), 
               columns = list(range(8)) + ["class"])
```

```python
splom_viz(df, 'class')
```

```python
from pandas.tools.plotting import radviz
def rad_viz(df,labels):
    fig = radviz(df, labels, color=sns.color_palette())
    plt.savefig("fea_rad_viz.png")
    plt.show()
```

```python
rad_viz(df, 'class')
```

```python
plt.figure(figsize=(12, 10))
_ = sns.linearmodels.corrplot(df, annot=False)
plt.savefig("fea_corr.png")
plt.savefig("fea_corr.svg")
plt.show()
```
